```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Users\triti\Downloads\AirPassengers.csv")
X = df['Passengers']
```

```python
plt.title('Airline Passengers Dataset', size = 20)
plt.plot(X)
```

```python
plot_acf(X)
```

The pattern is indicative of seasonality and cyclic components

```python
plot_pacf(X)
```

Three significant lags hint at the requirement of third order differencing. Here, we will display the results for second-order differencing.

```python
# First-order differencing
X_diff = X.diff(periods = 1)
X_diff = X_diff.dropna()
```

```python
# Plotting the differenced series
plt.title('Airline Passengers dataset after First-order differencing', size = 20)
plt.plot(X, label = 'Passengers')
plt.plot(X_diff, label = 'First-order Differenced Series', color = 'orange')
plt.legend()
```

```python
# Calculating autocorrelation values
acf_values = acf(X_diff)
acf_values
```

```python
# Plotting acf values
plot_acf(X_diff, lags = 30)
```

```python
# Calculate partial autocorrelation
pacf_values = pacf(X_diff)
pacf_values
```

```python
# Plot pacf values
plot_pacf(X_diff, lags=30)
```

```python
# Second-order differencing
X2 = X_diff.diff(periods = 1)
X2 = X_diff.dropna()

plot_acf(X2)
```

```python
# Plotting the differenced series
plt.title('Airline Passengers dataset after Second-order differencing', size = 20)
plt.plot(X, label = 'Passengers')
plt.plot(X2, label = 'Second-order Differenced Series', color = 'orange')
plt.legend()
```
