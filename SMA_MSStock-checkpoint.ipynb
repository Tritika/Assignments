## Smoothening Techniques: Simple Moving Average

Dataset: Microsoft Stock

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\triti\Downloads\Microsoft_Stock.csv")
```

```python
df.head()
```

```python
df['Open'].plot(color = 'green', figsize = (20,6))
```

## 3 Yearly Moving Average

```python
df['SMA_3'] = df['Close'].rolling(window = 3).mean()
print(df['SMA_3'].head(10))
```

```python
df['SMA_3'].plot(color = 'red', linewidth = 2, figsize = (20,6))
```

## 5 Yearly Moving Average

```python
df['SMA_5'] = df['Close'].rolling(window = 5).mean()
print(df['SMA_5'].head(10))
```

```python
df['SMA_5'].plot(color = 'orange', linewidth = 2, figsize = (20,6))
```

## 7 Yearly Moving Average

```python
df['SMA_7'] = df['Close'].rolling(window = 7).mean()
print(df['SMA_7'].head(10))
```

```python
df['SMA_7'].plot(color = 'blue', linewidth = 2, figsize = (20,6))
```

## 10 Yearly Moving Average

```python
df['SMA_10'] = df['Close'].rolling(window = 10).mean()
print(df['SMA_10'].head(10))
```

```python
df['SMA_10'].plot(color = 'black', linewidth = 2, figsize = (20,6))
```
