## Augmented Dickey - Fuller statistic to check stationarity

Null Hypothesis: The time series data is non-stationary

Alternative Hypothesis: The time series data is stationary


Decision Rule:

    - If the p-value is less than or equal to 0.05, you reject H0 and 
      conclude that the time series is stationary.
      
    - If the p-value is greater than 0.05 , you fail to reject H0 and 
      conclude that the time series is not stationary.

```python
t = train["Value"].values
result = adfuller(t)

print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

Result:

As you can see, the p-value is greater than 0.05. So we fail to reject the null hypothesis. The Time series is non-stationary.

## Data Transformation - From non stationary to stationary
### Hypotheses

H01: The time series data is non stationarity when log transformation is used.

H02: The time series data is non stationarity when power transformation used.

H11: The Time series data is Stationarity when Log transformation is used.

H12: The time series data is Stationarity when Power transformation is used.

```python
#Log transformation
X = np.log(train['Value'])
plt.plot(X)

result = adfuller(X)
print('ADF Statistic: %s' % result[0])
print('p-value: %s' % result[1])
```

```python
#Power transformation
train['Value_log'] = np.log(train['Value'])
train['Value_log_diff'] = train['Value_log'] - train['Value_log'].shift(1)

#Power transformation
Y = train['Value_log_diff'].dropna()
Y.plot()

result = adfuller(Y)
print('ADF Statistic: %s' % result[0])
print('p-value: %s' % result[1])
```

## Conclusion

1. Transforming the electric production dataset by log transformation we obtained ADF test p-value as 0.023, which is less than 0.05. Hence, we fail to accept the first null hypothesis.


2. Transforming the electric production dataset by power transformation we obtained ADF test p-value which is also less than 0.05. Hence, we fail to accept the second null hypothesis.


3. In this analysis, the data is transformed by both logarithmic and power transformations.
