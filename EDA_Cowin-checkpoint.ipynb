```python
import pandas as pd
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.offline as py
import plotly.express as px
import missingno as msno
sns.set_theme(style="darkgrid")
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'C:\Users\triti\Downloads\owid-covid-data.csv')
df.shape
```

```python
vcn = df.filter(['continent' , 'location', 'date', 'population',
       'population_density', 'total_vaccinations', 'people_vaccinated',
       'people_fully_vaccinated', 'Unnamed: 9',
       'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred',
       'people_fully_vaccinated_per_hundred', 'year', 'month'], axis=1)
vcn.head()
```

```python
vcn['year'] = pd.DatetimeIndex(vcn['date']).year
vcn.columns
```

```python
people_partly_vaccinated = df['people_vaccinated'] - df['people_fully_vaccinated']
vcn = vcn.assign(people_partly_vaccinated = people_partly_vaccinated)
```

## Data Cleaning

```python
#Checking the count and percentage of missing values in all columns
nmiss = vcn.isna().sum().sort_values(ascending = False)
missing_percentages = (vcn.isna().sum().sort_values(ascending = False)/len(vcn))*100

data = {'Count':nmiss,'Percentage':missing_percentages}
columns = vcn.columns

miss_df = pd.DataFrame(data, columns)
miss_df.sort_values(['Percentage'], ascending=False)
```

```python
# Using ffill and bfill for imputing missing values
for col in vcn.columns:
    vcn[col].ffill(axis = 0, inplace = True)
    vcn[col].bfill(axis = 0, inplace = True)
```

## Exploratory Data Analysis

#### What is the change in number of vaccinations across the world since 2019?

```python
plt.subplots(figsize=(15,6))
sns.countplot(x = 'year', data = vcn)
plt.xticks(rotation=90)
plt.title('Total number of vaccinations each year')
plt.show()
```

#### What is the precentage distribution of people fully vaccinated across all continents?

```python
ful = vcn.groupby('people_fully_vaccinated')['continent'].value_counts(normalize=True)
ful = ful.mul(100).rename('Percent').reset_index()
ful.head()
```

#### What is the distribution of people fully and partly vaccinated across all continents?

```python
cont = vcn.groupby('continent').agg({'people_partly_vaccinated':'sum','people_fully_vaccinated':'sum'})
cnt = cont.plot.barh(stacked=True)
plt.title('Vaccination count across continents')

fig=plt.gcf()
fig.set_size_inches(15,6)
plt.show()
```

#### Which countries in Asia have the highest number of people vaccinated per hundred?

```python
asia = vcn[vcn['continent'] == 'Asia']
top = pd.DataFrame(asia['location'].loc[vcn['people_vaccinated_per_hundred'] > 80 ])
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from os import path
from PIL import Image

text = " ".join(value for value in top.location.astype(str))
print ("There are {} words in the combination of all cells in the column location.".format(len(text)))

stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white", width=800, height=400).generate(text)

plt.axis("off")
fig=plt.gcf()
fig.set_size_inches(15,8)

plt.tight_layout(pad=0)
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```
